```python
#outlier treatment
def remove_outlier(df_revised,Quantity):
    Q1=df_revised[Quantity].quantile(0.25)
    Q3=df_revised[Quantity].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-(1.5*IQR)
    upper_bound=Q3+(1.5*IQR)
    return df_revised[(df_revised[Quantity]>=lower_bound) & (df_revised[Quantity]<=upper_bound)]
```

```python
columns=df_revised.select_dtypes(['int64','float64']).columns
columns
```

```python
df_revised_removed_outlier=remove_outlier(df_revised,"Quantity")
```
